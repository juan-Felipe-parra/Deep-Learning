### Sistema 1: 

1. \begin{equation}
\dot { x } =ax
\end{equation}
\begin{equation}
\dot { y } =-y
\end{equation}

```
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
```

```

x=np.linspace(-10,10,100)
r=-5
x_dot=1+r*x+x**2
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.spines['left'].set_position(('data',0.0))
ax.spines['bottom'].set_position(('data',0.0))
ax.spines['top'].set_color('none')
ax.spines['right'].set_color('none')

#plt.axis([-5, 5, -5, 5])
plt.plot(x,x_dot)
plt.grid()

x=sp.symbols('x')
x_dot1=1+r*x+x**2
roots=sp.solve(x_dot1)
print(roots)
for j in range(len(roots)):
  plt.plot(roots[j],0,'ro')
```

```
def fun (t,xy,a,b):
  x,y=xy
  return [a*x,-y]
  
a,b=-2,2

ti=0; tf=10; dt=1000
tspam=[ti, tf]
t=np.linspace(ti,tf,dt)
for x0 in np.arange(-5,5,1):
  for y0 in np.arange(-5,5,1):
    sol=solve_ivp(fun,tspam,[x0,y0],args=(a,b),t_eval=t)
    plt.plot(sol.y[0], sol.y[1])

plt.axis([-5, 5, -5, 5])
plt.grid()


```

```
ti=0; tf=100; dt=100
tspam=[ti, tf]

for r in np.arange(-5,5,0.2):
  for x0 in np.arange(-5,5,0.7):
    sol=solve_ivp(fun,tspam,[x0],args=(r,b))
    plt.scatter(r, sol.y[0][-1], color='darkgreen', marker='.')

tspam=[ti, -tf]

for r in np.arange(-5,5,0.2):
  for x0 in np.arange(-5,5,0.7):
    sol=solve_ivp(fun,tspam,[x0],args=(r,b))
    plt.scatter(r, sol.y[0][-1], color='red', marker='.')


plt.axis([-5, 5, -10, 10])
plt.grid()
```
